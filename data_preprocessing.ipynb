```python
import pandas as pd
import datetime
from importnb import Notebook
with Notebook():
    from utility import Utility

class Data:
    def __init__(self,path):
        self.data = pd.read_csv(path)
        self.data_cleaned = None
        
    def get_data_info(self,data):
        print("------------------------------First Five rows:------------------------------")
        print(data.head(5))
        print("------------------------------Last Five rows:------------------------------")
        print(data.tail(5))
        print("------------------------------Data Info:------------------------------")
        print(data.info())
        print("------------------------------Data Description------------------------------")
        print(data.describe().transpose())
    
    def clean_data(self):
        Todays_date = datetime.datetime(2020, 6, 15)
        
        util = Utility()
        
        
        self.data['TransactionTime'] = util.parse_date(self.data['TransactionTime'])
        self.data['TransactionTime'] = pd.to_datetime(self.data['TransactionTime'])
        
        self.data_cleaned = self.data[(self.data['UserId']>0)]                                         #ignoring the rows with UserID == -1
        self.data_cleaned = self.data_cleaned[(self.data_cleaned['ItemCode']>0)]                       #ignoring the rows with ItemCode == -1
        self.data_cleaned = self.data_cleaned[(self.data_cleaned['NumberOfItemsPurchased']>0)]         #ignoring the rows with NumberOfItemsPurchased < 0  
        self.data_cleaned = self.data_cleaned[(self.data_cleaned['CostPerItem']>0)]                    #ignoring the rows with CostPerItem < 0  
        self.data_cleaned = self.data_cleaned[(self.data_cleaned['TransactionTime']<Todays_date)]      #ignoring the rows with TransactionTime > Todays_date
        
        self.data_cleaned['TotalPrice'] = self.data_cleaned['NumberOfItemsPurchased'] * self.data_cleaned['CostPerItem']
        
        print('Number of rows ignored',(len(self.data)-len(self.data_cleaned)))
        
    
    
```

```python
transaction_data = Data('transaction_data.csv')
```

```python
transaction_data.get_data_info(transaction_data.data)
```

```python
transaction_data.clean_data()
```

```python
transaction_data.get_data_info(transaction_data.data_cleaned)
```
